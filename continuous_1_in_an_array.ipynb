## python code to find the length of all continuous 1 in a list  ## 
## compute the histogram ## 
## fit it to exponential function ## 
## this code can be used to find the survival probability ## 

```python
import numpy as np 
import random
```

```python
def find_length_of_ones(lst):
    max_lengths = []
    current_length = 0
    for item in lst:
        if item == 1:
              current_length += 1
        else:
            if current_length > 0:
                max_lengths.append(current_length)
            current_length = 0
  # Add the final sequence of 1's to the list of lengths
    if current_length > 0:
        max_lengths.append(current_length)
    return max_lengths
```

```python
# Generate a list of 100 random 0's and 1's
lst = [random.randint(0, 1) for _ in range(100)]

# Find the lengths of all continuous sequences of 1's in the list
lengths = find_length_of_ones(lst)

# Print the list of lengths
print(lengths)
print(lst)
```

```python
import random
import matplotlib.pyplot as plt

# Generate a list of 100 random 0's and 1's
lst = [random.randint(0, 1) for _ in range(100)]

# Find the lengths of all continuous sequences of 1's in the list
lengths = find_length_of_ones(lst)

# Plot the distribution of lengths using a histogram
plt.hist(lengths)
plt.xlabel('Length of continuous sequence of 1\'s')
plt.ylabel('Number of sequences')
plt.show()
```

```python
import random
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

# Generate a list of 100 random 0's and 1's
lst = [random.randint(0, 1) for _ in range(100)]

# Find the lengths of all continuous sequences of 1's in the list
lengths = find_length_of_ones(lst)

# Plot a histogram of the lengths using matplotlib
plt.hist(lengths, bins=max(lengths), density=True)

# Define the exponential function to fit the data
def exp_func(x, a, b):
    return a * np.exp(-b * x)

# Fit the histogram data to the exponential function
params, _ = curve_fit(exp_func, np.arange(1, max(lengths) + 1),
                       plt.hist(lengths, bins=max(lengths), density=True)[0])

# Plot the fitted curve on top of the histogram
plt.plot(np.arange(1, max(lengths) + 1),
         exp_func(np.arange(1, max(lengths) + 1), params[0], params[1]))

# Show the plot
plt.show()
```
